```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
COLUMNS = ['target', 'id', 'date', 'flag', 'user', 'text']
```

```python
df = pd.read_csv("./asset/training.1600000.processed.noemoticon.csv",  encoding="latin-1", header=None, names=COLUMNS)
```

```python
df.info
```

```python
# Le dataset encode : 0 = négatif, 4 = positif -> on passe en binaire
df['target'] = df['target'].replace(4, 1)
```

```python
df.head
```

```python
# Infos générales
print(df.info())
print(df.isnull().sum())
print(df.duplicated().sum())
```

```python
# Distribution des classes
print(df['target'].value_counts())
print(df['target'].value_counts(normalize=True))
```

```python
# Longueur des tweets
df['word_count'] = df['text'].apply(lambda x: len(str(x).split()))
print(df.groupby('target')['word_count'].describe())
```
