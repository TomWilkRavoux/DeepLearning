```python
# Analyse des tendances
print("=== TENDANCES GLOBALES ===")
print(df['sentiment_label'].value_counts())
print()
print(df['sentiment_label'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%')
```

```python
intervals = [(0, 0.1), (0.1, 0.2), (0.2, 0.3), (0.3, 0.4), (0.4, 0.5),
             (0.5, 0.6), (0.6, 0.7), (0.7, 0.8), (0.8, 0.9), (0.9, 1.0)]

for low, high in intervals:
    if high == 1.0:
        count = ((df["score_sentiment"] >= low) & (df["score_sentiment"] <= high)).sum()
    else:
        count = ((df["score_sentiment"] >= low) & (df["score_sentiment"] < high)).sum()
    print(f"{low}-{high} : {count}")
```
